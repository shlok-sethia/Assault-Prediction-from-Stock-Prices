```python
##Reading the csv file of Crime data

#final = pd.read_csv("boston_crime.csv")

final = pd.read_csv('cleaned_data.csv')

final['Crimes'] = 1 
final
```

```python
#Drop ID column
final = final.drop(columns=['INCIDENT_NUMBER'])
```

```python
#Performing column name editing for better readability 
final.columns = final.columns.str.replace(' ', '_')
final
```

```python
# Getting the stocks data from yahoo

# Companies majorly operated/headquarted in Boston City Area are only considered

stocks = ['BSX', 'RTX', 'MPGSX','STT','GE','EIX','TJX','JHI','TMO','VRTX','ADI','BCG','SPY']

import yfinance as yf
import pandas_datareader.data as pdr

yf.pdr_override()

#Start Date and End Dates to retreive the stock data for
start = "2018-01-01"
end = "2021-11-19"

closing_df = pdr.get_data_yahoo(stocks, data_source='yahoo', start=start, end=end)['Adj Close']
closing_df
```

```python
#Converting to DateTime data type
final['DATE'] = pd.to_datetime(final['DATE'])

#Merge Boston crime data with stock data of selected cities for each day 
final = final.merge(closing_df,how='left',left_on='DATE',right_on='Date')

for i in stocks:
    final[i].fillna(method='bfill',inplace=True)
```

```python
##Clustering 01 ~ Calendar attributes

Calendar_Crimes = final.groupby(['MONTH', 'HOUR','DAY_OF_WEEK','WEEKEND','DAY_OF_MONTH']).agg({'Crimes': ['sum']})
#Calendar_Crimes = final.groupby([ 'HOUR','DAY_OF_MONTH']).agg({'Crimes': ['sum']})

Calendar_Crimes.columns = ['TotalCrimes']
Calendar_Crimes = Calendar_Crimes.reset_index()
# print(Calendar_Crimes)
# np.exp(Model.params)
print(np.mean(Calendar_Crimes.TotalCrimes))

import pandas as pd
from statsmodels.formula.api import poisson
import numpy as np
import scipy.special as spsp
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.base.model import GenericLikelihoodModel
import scipy.stats as spst
from sklearn.cluster import KMeans 

n_cluster=3
model=KMeans(n_clusters=n_cluster)
model.fit(Calendar_Crimes.iloc[:,0:5])
clusters=model.predict(Calendar_Crimes.iloc[:,0:5])
Calendar_Crimes=Calendar_Crimes.assign(clusters=clusters)
Calendar_Crimes

np.round(model.cluster_centers_,6)
print(Calendar_Crimes.groupby("clusters").mean())
plt.scatter(Calendar_Crimes.HOUR,Calendar_Crimes.DAY_OF_MONTH, c=Calendar_Crimes.clusters)
```

```python
print("The cluster groups didn't make sense as the values of calendar attributes are fixed (on a known band scale). All the data points have only these values. But different insights are identified from this below")
```

```python
print("Findings - ")
print("1. Hour of the day at which the crime occured is clearly differentiating the number of crimes. Crimes are low in the early morning hours(not a favourite time for activists), and highest in the evenings post business hours")

```

```python
#Running separate poisson model for each cluster
print("Each model now has separate and more accurate coefficients/intercepts to represent the segments in clustered data")

for i in range(n_cluster):
    model=poisson("TotalCrimes~1", Calendar_Crimes[Calendar_Crimes.clusters==i]).fit(disp=0)
    print(np.exp(model.params.values))
```

```python
## adding covariates to Poisson distribution 
#MONTH + HOUR +  DAY_OF_WEEK+ WEEKEND
model = poisson("TotalCrimes ~ MONTH + HOUR +  DAY_OF_WEEK+ WEEKEND ", Calendar_Crimes).fit()
print(np.exp(model.params))
model.summary()
```

```python
print("Weekend has a negative coefficient. It means, crimes are lesser on the weekend")
```

```python
##Clustering 02 ~ Weather attributes

Weather_Crimes = final.groupby(['Temperature', 'Precipitation','Snow','Snow_Depth','Wind_Speed','Visibility','Cloud_Cover','Relative_Humidity']).agg({'Crimes': ['sum']})
Weather_Crimes.columns = ['TotalCrimes']
#Weather_Crimes = Weather_Crimes.rename(columns={'Cloud Cover': 'Cloud_Cover'})
Weather_Crimes = Weather_Crimes.reset_index()
# print(Weather_Crimes)
print(np.mean(Weather_Crimes.TotalCrimes))


import pandas as pd
from statsmodels.formula.api import poisson
import numpy as np
import scipy.special as spsp
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.base.model import GenericLikelihoodModel
import scipy.stats as spst
from sklearn.cluster import KMeans 

n_cluster=5
model=KMeans(n_clusters=n_cluster)
model.fit(Weather_Crimes.iloc[:,0:8])
clusters=model.predict(Weather_Crimes.iloc[:,0:8])
Weather_Crimes=Weather_Crimes.assign(clusters=clusters)
Weather_Crimes

np.round(model.cluster_centers_,6)
print(Weather_Crimes.groupby("clusters").mean())
# plt.scatter(Weather_Crimes.Temperature,Weather_Crimes.Relative_Humidity, c=Weather_Crimes.clusters)
plt.scatter(Weather_Crimes.Temperature,Weather_Crimes.Cloud_Cover, c=Weather_Crimes.clusters)
```

```python
#Checking the cluster means to find out the uniqueness
Weather_Crimes.groupby("clusters").mean()
```

```python
print("Scatter Plot shows 5 clusters - ")
print("1. High temperature / Summer days and High cloud cover ")
print("2. High temperature / Summer days and Low cloud cover ")
print("3. Medium temperature days")
print("4. Low temperature / Winter days and High cloud cover ")
print("5. Low temperature / Winter days and Low cloud cover ")
```
