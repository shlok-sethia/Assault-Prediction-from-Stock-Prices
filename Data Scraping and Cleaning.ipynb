```python
#reading the data from the csv file

data = pd.read_csv('boston_crime.csv',encoding='latin')
```

```python
#checking the number of non-null values and the data type

data.info()
```

```python
data = data.sort_values(by='OCCURRED_ON_DATE')
```

```python
data
```

```python
#dropping the rows will null values in the following columns

data = data.dropna(subset=['REPORTING_AREA'])
data = data.dropna(subset=['DISTRICT'])
data = data.dropna(subset=['Lat'])
data = data.dropna(subset=['STREET'])
data = data.drop(columns=['OFFENSE_CODE_GROUP'])
data = data.drop(columns=['DAY_OF_WEEK'])
```

```python
data.info()
```

```python
#checking the values present in the'SHOOTING' column
data['SHOOTING'].unique()
```

```python
#filling the null values with shooting=0
data['SHOOTING'] = data['SHOOTING'].fillna(0)
```

```python
#replacing values appropriately to convert 'SHOOTING' into a 0,1 column
data['SHOOTING'].replace({"0": 0, "1": 1, "Y":1}, inplace=True)
```

```python
data['SHOOTING'].unique()
```

```python
data.info()
```

```python
#converting the 'OCCURRED_ON_DATE to 'datetime' type to extract useful information

data['OCCURRED_ON_DATE'] = pd.to_datetime(data['OCCURRED_ON_DATE'], format='%Y-%m-%d %H:%M:%S')
data['DATE'] = data['OCCURRED_ON_DATE'].apply(lambda x:x.date())
data['DATE'] = pd.to_datetime(data['DATE'], format='%Y-%m-%d')
```

```python
data.info()
```

```python
data
```

```python
#extracting useful information like day of the week, whether it is a weekend or not, and finally the day of the month

data['DAY_OF_WEEK'] = data['OCCURRED_ON_DATE'].apply(lambda x:x.isoweekday())
data['WEEKEND'] = data['DAY_OF_WEEK'].apply(lambda x: 0 if x<6 else 1)
data['DAY_OF_MONTH'] = data['OCCURRED_ON_DATE'].apply(lambda x:x.strftime("%d"))
```

```python
data
```

```python
#cleaning the 'OFFENSE_DESCRIPTION' column by reducing the crime division from 800+ to 15 main categories and creating a new column called 'race_label'

def label_race (row):
    if 'ASSAULT' in row['OFFENSE_DESCRIPTION']:
        return 'ASSAULT'
    elif 'LARCENY' in row['OFFENSE_DESCRIPTION']:
        return 'LARCENY'
    elif 'M/V' in row['OFFENSE_DESCRIPTION']:
        return 'ACCIDENTS'
    elif 'DRUG' in row['OFFENSE_DESCRIPTION']:
        return 'DRUGS'
    elif 'FRAUD' in row['OFFENSE_DESCRIPTION']:
        return 'FRAUD'
    elif 'THREAT' in row['OFFENSE_DESCRIPTION']:
        return 'THREAT'
    elif 'AUTO' in row['OFFENSE_DESCRIPTION']:
        return 'AUTO THEFT'
    elif 'MANSLAUGHTER' in row['OFFENSE_DESCRIPTION']:
        return 'MURDER'
    elif 'PROSTITUTION' in row['OFFENSE_DESCRIPTION']:
        return 'PROSTITUTION'
    elif 'RAPE' in row['OFFENSE_DESCRIPTION']:
        return 'RAPE'
    elif 'ROBBERY' in row['OFFENSE_DESCRIPTION']:
        return 'ROBBERY'
    elif 'BURGLARY' in row['OFFENSE_DESCRIPTION']:
        return 'BURGLARY'
    elif 'B&E' in row['OFFENSE_DESCRIPTION']:
        return 'BURGLARY'
    elif 'KIDNAPPING' in row['OFFENSE_DESCRIPTION']:
        return 'KIDNAPPING'
    elif 'ARSON' in row['OFFENSE_DESCRIPTION']:
        return 'FIRE/ARSON'
    elif 'FIRE ' in row['OFFENSE_DESCRIPTION']:
        return 'FIRE/ARSON'
    elif 'VANDALISM' in row['OFFENSE_DESCRIPTION']:
        return 'VANDALISM'
    elif 'GRAFFITI' in row['OFFENSE_DESCRIPTION']:
        return 'VANDALISM'
    
    return 'OTHER'


data['race_label'] = data.apply (lambda row: label_race(row), axis=1)
```

```python
#further cleaning 

final = final.replace(' ',np.nan)
final = final[final['REPORTING_AREA'].notna()]
final['REPORTING_AREA'] = final['REPORTING_AREA'].astype(float)
final['REPORTING_AREA'] = final['REPORTING_AREA'].astype(int)
```

```python
cleaned_date = pd.DataFrame(final.groupby(['DATE'])['Snow'].count())
cleaned_date = cleaned_date.reset_index()
cleaned_date = cleaned_date.rename(columns={'Snow': "count"})
df4 = final[['DATE','Temperature','Visibility','Cloud Cover','Relative Humidity','Conditions','Close*',\
          'Precipitation','Snow','Snow Depth','Wind Speed','DAY_OF_WEEK','WEEKEND','DAY_OF_MONTH',\
          'YEAR','MONTH','UNEMPLOY_RATE','FEDERAL HOLIDAY','GOVERNMENT HOLIDAY','NON-PUBLIC HOLIDAY','REGIONAL HOLIDAY','race_label']]
df4 = df4.drop_duplicates()

cleaned_date = cleaned_date.merge(df4,how='left',left_on='DATE',right_on='DATE')
cleaned_date = cleaned_date.rename(columns={"Close*":'DJI','Relative Humidity':'humidity',"Snow Depth":"snow_depth"})
cleaned_date = cleaned_date.rename(columns={"Wind Speed":'wind_speed','FEDERAL HOLIDAY':'fed_holiday','GOVERNMENT HOLIDAY':'gov_holiday'})
cleaned_date = cleaned_date.rename(columns={"NON-PUBLIC HOLIDAY":'nonpub_holiday','REGIONAL HOLIDAY':'reg_holiday'})
```

```python
stocks = ['JHI','MPGSX','TJX','TMO','BSX','EIX','STT','BCG']

yf.pdr_override()


start = "2018-01-01"
end = "2021-11-19"

closing_df = pdr.get_data_yahoo(stocks, data_source='yahoo', start=start, end=end)['Adj Close']
```

```python
cleaned_date['DATE'] = pd.to_datetime(cleaned_date['DATE'])
cleaned_date = cleaned_date.merge(closing_df,how='left',left_on='DATE',right_on='Date')
```

```python
for i in stocks:
    cleaned_date[i].fillna(method='bfill',inplace=True)

predictors = ['UNEMPLOY_RATE']
predictors.extend(stocks)  

### ROLLING AVERAGES

for i in predictors:
    cleaned_date[str(i)+'_7'] = cleaned_date[i].rolling(window=7).mean()
```

```python
#creating lag variables. We are predicting the number of assaults happening in the next three days, hence we create a shift of 3

lag_columns = ['count','Temperature','DJI','UNEMPLOY_RATE','Conditions','Precipitation','Snow']
lag_columns.extend(['JHI','MPGSX','TJX','TMO','JHI_7','MPGSX_7','TJX_7','TMO_7','BCG_7'])

for i in lag_columns:
    cleaned_date[str(i)+'_lag'] = cleaned_date[i].shift(3)

cleaned_date = cleaned_date.assign(t = lambda x:x.index)
```
